# Bivariate Analysis

```python
df['mrp']
```

```python
def remove_char(val):
    # Remove the currency symbol and any non-numeric characters
    cleaned_val = ''.join(filter(str.isdigit, val))
    return int(cleaned_val)
```

```python
df['mrp'] = df['mrp'].map(remove_char)
```

```python
df.mrp
```

```python
df.selling_price=df.selling_price.map(remove_char)
```

```python
sns.boxplot(x='category_1',y='product_rating',data=df)
plt.title('Category_1 vs Product Rating\n')
plt.figure(figsize=(10,10))
plt.xticks(rotation=45)
plt.show()
```

```python
sns.jointplot(x='product_rating',y='selling_price',data=df)
plt.show()
```

```python
product_selling_price = (
    df.groupby(['category_3'])['selling_price']
    .agg(max)
    .sort_values(ascending=False)[:10]
)
```

```python
product_selling_price
```

```python
plt.figure(figsize=(15, 8))

# Get the top 10 maximum selling prices for each category
product_selling_price = (
    df.groupby(['category_3'])['selling_price']
    .agg(max)
    .sort_values(ascending=False)[:10]
)

# Create a bar plot
G = sns.barplot(x=product_selling_price.index, y=product_selling_price.values, palette='magma')

# Annotate the bars with the values
for i in range(10):
    value = product_selling_price.values[i]
    G.text(y=value - 2, x=i, s=value, color='black', ha='center', fontsize=15)

# Set labels and title
G.set_xlabel('Product', fontsize=15)
G.set_ylabel('Selling Price', fontsize=15)
G.set_title('Top 10 Products with Highest Selling Price\n', color='darkred', fontsize=30)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
plt.show()
```

```python
# Get the top 10 maximum MRP values for each category
vc = df.groupby(['category_3'])['mrp'].agg(max).sort_values(ascending=False)[:10]

plt.figure(figsize=(15, 10))

# Create a bar plot
G = sns.barplot(x=vc.index, y=vc.values, palette='Set2')

# Annotate the bars with the values
for i in range(10):
    value = vc.values[i]
    G.text(y=value - 2, x=i, s=value, color='black', ha='center', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Products', fontsize=12)
plt.ylabel('MRP', fontsize=15)
plt.title('Top 10 Products with Highest MRP', fontsize=20)

plt.show()
```

```python
# Get the top 10 maximum product ratings for each category
vc = df.groupby(['category_3'])['product_rating'].agg(max).sort_values(ascending=False)[:10]

plt.figure(figsize=(12, 5))

# Create a bar plot
G = sns.barplot(x=vc.index, y=vc.values, palette='Set2')

# Annotate the bars with the values
for i in range(10):
    value = vc.values[i]
    G.text(y=value - 0.1, x=i, s=round(value, 2), color='black', ha='center', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Products', fontsize=12)
plt.ylabel('Ratings', fontsize=15)
plt.title('Top 10 Products as per Ratings', fontsize=20)

plt.show()
```

```python
# Get the top 10 maximum seller ratings for each seller
vc = df.groupby(['seller_name'])['seller_rating'].agg(max).sort_values(ascending=False)[:10]

plt.figure(figsize=(12, 5))

# Create a bar plot
G = sns.barplot(x=vc.index, y=vc.values, palette='Set2')

# Annotate the bars with the values
for i in range(10):
    value = vc.values[i]
    G.text(y=value - 0.1, x=i, s=round(value, 2), color='black', ha='center', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Seller', fontsize=12)
plt.ylabel('Ratings', fontsize=15)
plt.title('Top 10 Sellers as per Ratings', fontsize=20)

plt.show()
```
